```python
# Label map
label_map = {'Viable': 0, 'Non-Viable-Tumor': 1, 'Non-Tumor': 2}

output_folders = ['output_1', 'output_2']
data = []

for output_folder in output_folders:
    for class_label in os.listdir(output_folder):
        class_dir = os.path.join(output_folder, class_label)
        if os.path.isdir(class_dir):
            for img_name in os.listdir(class_dir):
                if img_name.lower().endswith(('.jpg')):
                    img_path = os.path.join(class_dir, img_name) #To build the image path
                    data.append({
                        'image_path': img_path.replace("\\", "/"),
                        'label': label_map[class_label]  # convert labels to integer
                    })
```

```python
#IMPORTING LIBRARIES AND PACKAGES
```

```python
import os
import pandas as pd
from PIL import Image  #Python Imaging library - to open & manipulate images
from tqdm import tqdm  #Progress bar
import torch  #Pytorch library for dl tasks
import torch.nn as nn #Contains all Neural network layers like conv2D, Relu, etc
import torch.optim as optim #Optimizers like Adam
from torchvision import transforms, models #transforms - image resize, normalize, augment; models - pretrained cnn models like googlenet
from torch.utils.data import Dataset, DataLoader #Dataset - to convert images in such a way that pytorch can handle it. DataLoader - Send images as mini-batches to the model for training.
from sklearn.model_selection import train_test_split
```

```python
# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu") #use GPU if available
```

```python
class OsteoSarcomaDataset(Dataset): #allows PyTorch’s DataLoader to load osteosarcoma images along with their labels from a DataFrame.
    def __init__(self, dataframe, transform=None):   #None because its optional
        self.df = dataframe 
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):     #tells PyTorch how to fetch one item (image + label) at the index idx
        image_path = self.df.iloc[idx]['image_path']
        label = int(self.df.iloc[idx]['label'])  # ensure it's an int
        image = Image.open(image_path).convert('RGB')
        if self.transform:             #applies transformations if given
            image = self.transform(image)
        return image, label
```

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])    #Compose - helps to merge multiple image transformations together
```

```python
#DATA SPLITTING
```

```python
train_df, test_df = train_test_split(combined_df, test_size=0.2, stratify=combined_df['label'], random_state=42) #stratify - to balance both train and test based on labels
```

```python
#Creating instances
train_dataset = OsteoSarcomaDataset(train_df, transform=transform)
test_dataset = OsteoSarcomaDataset(test_df, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
```

```python
#LOAD GOOGLENET
```

```python
model = models.googlenet(weights=models.GoogLeNet_Weights.DEFAULT)
model.fc = nn.Linear(model.fc.in_features, 3)  # 3 classes; fc layer by default outputs for 1000 classes
model = model.to(device) #gpu/cpu usage
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)
```

```python
#MODEL TRAINING
```

```python
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0 #loss of training set
    correct = 0  #correctly predicted
    total = 0   #no. of samples in the current batch
    for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}"):
        images, labels = images.to(device), labels.to(device)  #to gpu

        optimizer.zero_grad() 
        outputs = model(images)
        loss = criterion(outputs, labels)

        loss.backward()  #Backpropagation
        optimizer.step() #Update weights

        running_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1) #to identify the class of the sample
        total += labels.size(0) 
        correct += (predicted == labels).sum().item()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss:.4f}, Accuracy: {100 * correct / total:.2f}%")
```

```python
#MODEL EVALUATION
```

```python
model.eval()
correct = 0
total = 0

with torch.no_grad():  #no updation for test
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"\nTest Accuracy: {100 * correct / total:.2f}%")
```

```python
class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer #the layer to be focused for calculating gradients; last conv layer
        self.gradients = None #gradients from the backward pass
        self.activations = None #feature maps from the forward pass
        self.hook_layers() 
        
#hook layers - To combine features and gradients of target layer for grad cam application
    def hook_layers(self):
        def forward_hook(module, input, output):
            self.activations = output.detach() #detach- no more features needed(like a screenshot-can be viewed, used later).

        def backward_hook(module, grad_input, grad_output):
            self.gradients = grad_output[0].detach()  #grad_output - gradient of output of target layer
#registering hooks - allows to connect the hooks with forward and backward propagation code. It grabs the features without changing anything.
        self.target_layer.register_forward_hook(forward_hook)
        self.target_layer.register_backward_hook(backward_hook)

    def generate(self, input_tensor, class_idx=None):
        output = self.model(input_tensor) #forward pass; image through forward pass to predict output
        if class_idx is None:
            class_idx = output.argmax(dim=1).item() #

        self.model.zero_grad()
        class_score = output[0, class_idx]
        class_score.backward()

        pooled_gradients = torch.mean(self.gradients, dim=[0, 2, 3]) #average pooling
        activations = self.activations[0]  
        for i in range(activations.shape[0]): #takes activation maps and weights them 
            activations[i, :, :] *= pooled_gradients[i]

        heatmap = torch.mean(activations, dim=0).cpu().numpy() #Averages all channels of the weighted feature maps into a single 2D heatmap.
        heatmap = np.maximum(heatmap, 0) #Sets negative values to zero.Normalizes the heatmap so values range between 0 and 1.

        heatmap /= np.max(heatmap)

        return heatmap, class_idx
```

```python
def apply_gradcam(image_path, model, gradcam):
    # Load and preprocess image
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
    ])
    image = Image.open(image_path).convert("RGB")
    input_tensor = transform(image).unsqueeze(0).to(device)

    # Generate heatmap
    heatmap, pred_class = gradcam.generate(input_tensor)

    # Load image for visualization
    img = cv2.imread(image_path)
    img = cv2.resize(img, (224, 224))

    # Overlay heatmap
    heatmap = cv2.resize(heatmap, (224, 224))
    heatmap = np.uint8(255 * heatmap)
    heatmap_color = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    overlayed = cv2.addWeighted(img, 0.6, heatmap_color, 0.4, 0) #blending 0.6 weights of original image and 0.4 of heatmap.

    # Plot
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.title(f"Original")
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) #colored original image
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title(f"Grad-CAM (Predicted Class: {pred_class})") 
    plt.imshow(cv2.cvtColor(overlayed, cv2.COLOR_BGR2RGB)) 
    plt.axis('off')
    plt.show()
```

```python
def apply_gradcam_diff_btw_classes(image_paths, model, gradcam):
    plt.figure(figsize=(10, 12))  # 3 rows, 2 columns

    for i, image_path in enumerate(image_paths):
        # Load and preprocess image
        transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
        ])
        image = Image.open(image_path).convert("RGB")
        input_tensor = transform(image).unsqueeze(0).to(device)

        # Generate heatmap
        heatmap, pred_class = gradcam.generate(input_tensor)

        # Load original image for OpenCV overlay
        img = cv2.imread(image_path)
        img = cv2.resize(img, (224, 224))

        # Create color heatmap overlay
        heatmap = cv2.resize(heatmap, (224, 224))
        heatmap = np.uint8(255 * heatmap)
        heatmap_color = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
        overlayed = cv2.addWeighted(img, 0.6, heatmap_color, 0.4, 0)

        # Original image
        plt.subplot(3, 2, 2 * i + 1)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(f"Original")
        plt.axis('off')

        # Grad-CAM image
        plt.subplot(3, 2, 2 * i + 2)
        plt.imshow(cv2.cvtColor(overlayed, cv2.COLOR_BGR2RGB))
        plt.title(f"Grad-CAM (Class: {pred_class})")
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
from sklearn.preprocessing import label_binarize

model.eval()
all_preds = []
all_labels = []
all_probs = []
```

```python
with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        probs = torch.softmax(outputs, dim=1)

        all_probs.append(probs.cpu().numpy())
        all_preds.append(torch.argmax(probs, dim=1).cpu().numpy())
        all_labels.append(labels.cpu().numpy())
```

```python
# Concatenate all batches
y_true = np.concatenate(all_labels)
y_pred = np.concatenate(all_preds)
y_prob = np.concatenate(all_probs)
```

```python
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_auc_score,
    roc_curve  #This was missing in your case
)
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Binarize labels for ROC-AUC (One-hot encoding)
n_classes = y_prob.shape[1]
y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))

# Plot ROC curve for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_prob[:, i])
    roc_auc[i] = roc_auc_score(y_true_bin[:, i], y_prob[:, i])

# Plot all ROC curves
plt.figure(figsize=(8, 6))
colors = ['red', 'green', 'blue']
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], color=colors[i], label=f"Class {i} (AUC = {roc_auc[i]:.2f})")

plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Multi-Class ROC Curve")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
